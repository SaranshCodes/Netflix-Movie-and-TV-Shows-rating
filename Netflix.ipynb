```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\saran\Desktop\netflixData.csv")
```

```python
df.head(5)
```

```python
df.info()
```

```python
df["Imdb Score"]=df["Imdb Score"].str.replace("/10","")
df["Imdb Score"]=pd.to_numeric(df["Imdb Score"],errors="coerce")
```

```python
df.info()
```

```python
df.rename(columns={'Release Date':"Release Year"},inplace=True)
```

```python
df["Release Year"]=df["Release Year"].astype("Int64")
```

```python
df["Date Added"]=pd.to_datetime(df["Date Added"],errors="coerce")
```

```python
df["Genres"]=df["Genres"].astype(str).str.split(',')
df["Genres"]=df["Genres"].apply(lambda x: [genre.strip() for genre in x])
df=df.explode("Genres")
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.head(10)
```

```python
df.drop(columns=["description"],inplace=True)
```

```python
df["title"]=df["title"].str.replace("#","")
```

```python
df.to_csv("cleaned_netflix.csv",index= False)
```
